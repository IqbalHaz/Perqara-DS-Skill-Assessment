### Time Series on Purchase Date

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
```

```python
df.info()
```

```python
df2 = df[:]
df2.loc[:,'purchase_date'] = df.loc[:,'order_purchase_timestamp'].dt.date
df2.loc[:,'freq'] = 1
purchase_per_date = df2.groupby('purchase_date').aggregate({'freq': 'sum'}).reset_index()
purchase_per_date.head(10)
```

```python
plt.figure(figsize=(10,4))
plt.plot(purchase_per_date.purchase_date, purchase_per_date.freq)
```

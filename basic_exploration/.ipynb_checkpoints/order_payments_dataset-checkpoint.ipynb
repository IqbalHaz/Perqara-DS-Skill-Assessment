# A Look into Order Payments Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/order_payments_dataset.csv')
df
```

### Payment Value Distribution

```python
plt.hist(df.payment_value, bins=20)
plt.title('payment value')
plt.show()
```

### Payment Sequential Count

```python
df.payment_sequential.value_counts().reset_index(name='counts')
```

### Payment Installments Count

```python
df.payment_installments.value_counts().reset_index(name='counts')
```

### Payment Type Count

```python
payment_count = df.payment_type.value_counts().reset_index(name='counts')
payment_count
```

```python
fig, ax = plt.subplots()

types = payment_count.payment_type.to_numpy()
y_pos = np.arange(len(types))
counts = payment_count.counts.to_numpy()

ax.barh(y_pos, counts, align='center')
ax.set_yticks(y_pos, labels=types)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('No. of Customer using certain Payment Type')
ax.set_title('Each Payment Type Count')

plt.show()
```

# A Look into Order Items Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/order_items_dataset.csv')
df
```

## Data Exploration

### Data Description

```python
df.info()
```

As can be seen above, no missing value from `order_items_dataset`. <br> `order_item_id` has int64 dtype instead of object (compared to other id column) due to its solely numeric value.

```python
df.describe()
```

```python
df.describe(include=[object])
```

### Order Id Duplicates

```python
df[df.order_id.duplicated()]
```

### Price Distribution

```python
plt.hist(df.price, bins=20)
plt.title('price')
plt.show()
```

### Freight Value Distribution

```python
plt.hist(df.freight_value, bins=20)
plt.title('freight_value')
plt.show()
```

### Order Item Id Count

```python
df.order_item_id.value_counts().reset_index(name='counts')
```

### Numerical Columns Correlation

```python
sns.heatmap(df[['order_item_id', 'price', 'freight_value']].corr())
plt.title('Correlation between Numerical Columns')
plt.show()
```
